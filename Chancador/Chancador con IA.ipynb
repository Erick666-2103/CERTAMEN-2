```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import joblib
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from functools import partial
from pandas.tseries.frequencies import to_offset

# Construir la ruta completa del archivo
file_path = os.path.join(
    'C:/Users/eamo8/Desktop/Notebook para repositorio/Chancador',
    'Chancador_Raw_Data.csv'
)

# Leer el archivo CSV
df1 = pd.read_csv(file_path)

# Obtener atributos del DataFrame
attrs = {
    'var': 'df1',
    'type': type(df1).__name__,
}

# Mostrar atributos y el DataFrame
print(attrs)
df1
#--------------------------------------
# Eliminar columnas no deseadas del DataFrame
columnas_a_eliminar = [
    'V. Cero.1',
    'Corriente.4',
    'V. Cero.2',
    'Corriente.2',
    'Corriente.3',
    'V. Cero',
    'Unnamed: 0',
    'Corriente.1',
]
df1 = df1.drop(columns=columnas_a_eliminar)

# Obtener atributos del DataFrame
attrs = {
    'var': 'df1',
    'type': type(df1).__name__,
}

# Mostrar atributos y el DataFrame resultante
print(attrs)
df1
#----------------------------------------
# Renombrar columnas del DataFrame
df1 = df1.rename(columns={
    'Salto Anillo': 'Salto Anillo A',
    'Unnamed: 9': 'Salto Anillo B',
    'Unnamed: 10': 'Salto Anillo C',
    'Unnamed: 11': 'Salto Anillo D',
})

# Obtener atributos del DataFrame
attrs = {
    'var': 'df1',
    'type': type(df1).__name__,
}

# Mostrar atributos y el DataFrame resultante
print(attrs)
df1
#------------------------------------------
# Convertir columna de tiempo a tipo datetime
df1['Tiempo'] = pd.to_datetime(df1['Tiempo'], errors='coerce')

# Lista de columnas a convertir a tipo numérico
columnas_numericas = [
    'Corriente',
    'T° Retorno',
    'T° socket Linner',
    'T° Excéntrica',
    'T° Alimentación',
    'T° Desc. C. Eje',
    'Salto Anillo A',
    'Salto Anillo B',
    'Salto Anillo C',
    'Salto Anillo D',
    'Nivel Taza',
    'Setting',
    'Pesómetro',
]

# Conversión a numérico con manejo de errores
for col in columnas_numericas:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')

# Obtener atributos del DataFrame
attrs = {
    'var': 'df1',
    'type': type(df1).__name__,
}

# Mostrar atributos y el DataFrame
print(attrs)
df1
#-------------------------------------------
# Eliminar filas que contienen valores nulos
df1 = df1.dropna(axis='index')

# Obtener atributos del DataFrame
attrs = {
    'var': 'df1',
    'type': type(df1).__name__,
}

# Mostrar atributos y el DataFrame
print(attrs)
df1
#-------------------------------------------
# Establecer 'Tiempo' como índice del DataFrame
df1.set_index('Tiempo', inplace=True)

# Obtener atributos del DataFrame
attrs = {
    'var': 'df1',
    'type': type(df1).__name__,
}

# Mostrar atributos y el DataFrame
print(attrs)
df1
#-------------------------------------------

# Función para redondear un timestamp hacia abajo según la frecuencia dada
def roundfunc(t, freq):
    freq = to_offset(freq)
    return pd.Timestamp((t.value // freq.delta.value) * freq.delta.value)

# Agrupar df1 por intervalos de 10 minutos usando la función de redondeo
# Note: Renamed df20 to df21 in attrs to match variable name for clarity
df21 = df1.groupby(partial(roundfunc, freq='10T'))

# Mostrar atributos del objeto agrupado
attrs = {
    'var': 'df21', # Changed from df20 to df21
    'type': type(df21).__name__,
}
print(attrs)

# Explorar grupos uno por uno
enum = enumerate(df21.groups)
nextel = next(enum)
print(f"Primer grupo: {nextel[0]} -> {nextel[1]}")
print(df21.get_group(nextel[1]))

nextel = next(enum)
print(f"Segundo grupo: {nextel[0]} -> {nextel[1]}")
print(df21.get_group(nextel[1]))
#-------------------------------------------
# Aplicar agregación de mediana a cada grupo
df21 = df21.aggregate('median','object')

# Obtener atributos del nuevo DataFrame resultante
attrs = {
    'var': 'df21',
    'type': type(df21).__name__,
}

# Mostrar atributos y el DataFrame resultante
print(attrs)
df21
#-------------------------------------------
import os

# Construir la ruta completa del archivo CSV
file_path = os.path.join(
    'C:/Users/eamo8/Desktop/Notebook para repositorio/Chancador',
    'intento 5.csv'
)

# Guardar el DataFrame como archivo CSV
# Corrected line: removed index=False to include the 'Tiempo' index as a column
df21.to_csv(file_path)

# Establecer atributos (vacíos, como en tu ejemplo)
attrs = {
    'var': 'empty',
    'type': 'empty',
}

# Mostrar atributos y mensaje de éxito
print(attrs)
print('CSV file saved at:', file_path)
```
